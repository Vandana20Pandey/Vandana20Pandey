## Summary of the dataset

## Column Description
- **show_id** : Shows the unique show id
- **type** : Tells whether the content is a movie or a tv show
- **title** : Name of the show or movie
- **director** : Name of the director of the show or content
- **cast** : Name of the cast of the movie or tv show
- **country** : Name of the country in which content was produced
- **date added** : Date on which the content was added on netflix
- **release_year** : Release year of the content
- **rating** : rating provided to the content by authorising committee
- **duration**: Tells the total duration of the content
- **listed_in** : In which genres the content was listed in
- **description**: Overall description of the show or movie

## Sentiment Analysis of netflix content

```python
df3 = df[['release_year', 'description']]
df3.head()
```

```python
for index , row in df3.iterrows():
  d=row['description']
  testimonial = TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent = 'Neutral'
  elif p>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index,2], 'Sentiment']=sent
```

```python
df3 = df3.groupby(['release_year', 'Sentiment']).size().reset_index(name='count')
```

```python
df3 = df3[df3['release_year']>2005]
df3.head()
```

```python
px.bar(df3, x='release_year', y='count', color='Sentiment', title='Sentiment Analysis')
```

### **Conclusion** :- We can see that every year more positve content is streamed on netflix as compared to negative content
